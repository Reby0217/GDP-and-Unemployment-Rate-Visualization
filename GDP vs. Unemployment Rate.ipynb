**The GDP per cap and unempolyment rate for all the countries in the file are chosen because I need these information sources to investigate the relationship between these two variables. The GDP per cap is the y-axis of the scatterplot and the unemployment rate is the x-axis of the plot. I also need the unemployment rates of all the countries to calculate the average unemployment rate, which helps me filter the data (only the countrys unemployment rate is greater than the average, will the country's unemployment rate and GDP per cap values be plotted in the scatterplot ).**

```python
from cs103 import *
from typing import NamedTuple, List
import csv
import matplotlib.pyplot as plt

##################
# Data Definitions

CountryData = NamedTuple("CountryData", [("ur", float),       # in range [0.0, 100.0]
                                         ("gdp", int)])       # in range [0, ...)

#interp. a country's data with the country's unemployment rate ("ur") and it's gdp per cap ("gdp") in USD

CD0 = CountryData(19.3, 11305)
CD1 = CountryData(11.2, 14687)
CD2 = CountryData(11.0, 22968)
CD3 = CountryData(18.5, 8070)
CD4 = CountryData(5.8, 45514)
CD5 = CountryData(9.1, 47824)
CD6 = CountryData(22.6, 2182)
CD7 = CountryData(35.0, 10427)
CD8 = CountryData(17.8, 15807)
CD9 = CountryData(60.0, 3270)


@typecheck
def fn_for_country_data (cd: CountryData) -> ...:      # template based on Compound
    return ...(cd.ur,
               cd.gdp)




# List[CountryData]
# interp. a list of country data

LOCD0 = []
LOCD1 = [CD0, CD1]
LOCD2 = [CD0, CD1, CD2, CD3, CD4]

@typecheck
# template based on arbitrary-sized and reference rule
def fn_for_locd(locd: List[CountryData]) -> ...:
    # description of the accumulator
    acc = ...      # type: ...
    for cd in locd:
        acc = ...(fn_for_country_data(cd), acc)
 
    return ...(acc)




# List[float]
# interp. a list of floats, which are the unemployment rates

LOF0 = []
LOF1 = [19.3, 11.2]
LOF2 = [19.3, 11.2, 11.0, 18.5, 5.8]

@typecheck
def fn_for_lof(lof: List[float]) -> ...:     #template based on arbitrary-sized
    # description of the accumulator
    acc = ...      # type: ...
    for f in lof:
        acc = ...(f, acc)

    return ...(acc)




# List[int]
# interp. a list of ints, which are the gdp per cap values

LOI0 = []
LOI1 = [11305, 14687]
LOI2 = [11305, 14687, 22968, 8070, 45514]

@typecheck
def fn_for_loi(loi: List[int]) -> ...:       #template based on arbitrary-sized
    # description of the accumulator
    acc = ...      # type: ...
    for i in loi:
        acc = ...(i, acc)

    return ...(acc)

```

```python
###########
# Functions

@typecheck
def main(filename: str) -> None:
    """
    Reads the file from given filename, analyzes the data, plot a scatterplot of 
    GDP per cap (in USD) vs. unemployment rate for countries from locd with 
    unemployment rates greater than the average unemployment rate, and returns None 
    """
    # template from HtDAP, based on function composition 
    return plot_gdp_over_ur(read(filename)) 
    
    


@typecheck
def read(filename: str) -> List[CountryData]:
    """    
    reads information from the specified file and returns a list of country data
    """
    #return []  #stub
    # template from HtDAP
    # locd contains the result so far
    locd = [] # type: List[CountryData]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            cd = CountryData(parse_float(row[2]), parse_int(row[3]))
            locd.append(cd)
    
    return locd


@typecheck
def get_ur_list(locd: List[CountryData]) -> List[float]:
    '''
    return a list of floats, which are the unemployment rates from locd
    '''
    #return [] #stub
    # template from List[CountryData]
    # acc stores the unemployment rates from locd seen so far
    ur_list = []      # type: List[float]
    for cd in locd:
        ur_list.append(cd.ur)
 
    return ur_list



@typecheck
def get_gdp_list(locd: List[CountryData]) ->List[int]:
    '''
    return a list of ints, which are the gdp per cap in USD from the locd
    '''
    #return [] #stub
    # template from List[CountryData]
    # acc stores the gdp per cap values from locd seen so far
    gdp_list = []      # type: List[float]
    for cd in locd:
        gdp_list.append(cd.gdp)
 
    return gdp_list



@typecheck
def sum_of_ur(lof: List[float]) -> float:
    '''
    return the sum of floats, which are the unemployment rates, from lof
    '''
    #return 10.5 #stub
    #template from List[float]
    # ur_sum stores the sum of floats, which are the unemployment rates, from lof seen so far
    ur_sum = 0.0      # type: float
    for f in lof:
        ur_sum += f

    return ur_sum



@typecheck
def num_of_ur(lof: List[float]) -> int:
    '''
    return an int, which is the number of unemployment rate values from lof
    '''
    #return 0 #stub
    # template from List[float]
    #number stores the number of unemployment rate values from lof seen so far
    number = 0   # type: int
    for f in lof:
        number += 1

    return number



@typecheck
def avg_ur(lof: List[float]) -> float:
    '''
    return a float, which is the average value of the umemployment rates from lof
    if the number of of unemployment rate values from lof is 0 (lof is empty), return 0
    '''
    #return 5.4 #stub
    #template from List[float]
    if num_of_ur(lof) == 0:
        average = 0
    else:
        average = sum_of_ur(lof)/num_of_ur(lof)

    return average



@typecheck
def avg_ur_of_ur_list(locd: List[CountryData]) -> float:
    '''
    return a float, which is the average value of the umemployment rates from locd
    '''
    #return 0.0 #stub
    #template based on the composition rule
    return avg_ur(get_ur_list(locd))
    


@typecheck
def is_greater(f1:float , f2:float) -> bool:
    '''
    return True is f1 is greater than f2, False otherwise
    '''
    #return True
    #return...(f1, f2) #template
    return f1 > f2


@typecheck
def is_greater_than_avg(cd: CountryData, locd: List[CountryData]) -> bool:
    '''
    return True if the unemployment rate of the country's data is greater than 
    the average unemployment rate of the locd, False otherwise
    '''
    #return True #stub
    #template based on the composition rule
    return is_greater(cd.ur, avg_ur_of_ur_list(locd))
    


@typecheck
def filter_the_data(locd: List[CountryData]) -> List[CountryData]:
    '''
    return a list of countries' data from locd, whose unemployment rates are greater than the 
    average unemployment rate 
    ''' 
    #return [] #stub
    #template from List[CountryData]
    # acc stores the countries' data from locd seen so far, whose unemployment rates are 
    # greater than the average unemployment rate 
    acc = []      # type:List[CountryData]
    for cd in locd:
        if is_greater_than_avg(cd, locd):
            acc.append(cd)
 
    return acc



@typecheck
def plot_gdp_over_ur(locd: List[CountryData]) -> None:
    '''
    Display a scatterplot of GDP per cap (in USD) vs. unemployment rate for countries from locd with 
    unemployment rates greater than the average unemployment rate. 
    
    Assumes that the lengths of the list of GDP per cap values and the list of unemployment rates are equal.
    '''
    #return None #stub
    # template based on visualization
        # set the labels for the axes
    
    plt.xlabel("Unemployment rate")
    plt.ylabel('GDP per cap (in USD)')
    plt.title("GDP per cap over unemployment rate")

    # range for the axes
    # [x-min, x-max, y-min, y-max]
    plt.axis([0,100,0,40000])

    plt.scatter(get_ur_list(filter_the_data(locd)),get_gdp_list(filter_the_data(locd)),marker='o', c='r')

    # show the plot
    plt.show()
    
    return None




start_testing()

# Examples and tests for main
expect(main("country_per_cap_gdp_unemployment_gov_type_pop_test1.csv"), None)
expect(main("country_per_cap_gdp_unemployment_gov_type_pop_test2.csv"), None)


# Examples and tests for read
expect(read("country_per_cap_gdp_unemployment_gov_type_pop_test1.csv"), LOCD1)
expect(read("country_per_cap_gdp_unemployment_gov_type_pop_test2.csv"), LOCD2)


# Examples and tests for get_ur_list
expect(get_ur_list([]),[])
expect(get_ur_list(LOCD1),[19.3, 11.2])
expect(get_ur_list(LOCD2), LOF2)
expect(get_ur_list([CD0,CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8]), [19.3, 11.2, 11.0, 18.5, 5.8, 9.1, 22.6, 35.0, 17.8])


# Examples and tests for get_gdp_list
expect(get_gdp_list([]),[])
expect(get_gdp_list(LOCD1),[11305, 14687])
expect(get_gdp_list(LOCD2), LOI2)
expect(get_gdp_list([CD0,CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8]), [11305, 14687, 22968, 8070, 45514, 47824, 2182, 10427, 15807])


# Examples and tests for sum_of_ur
expect(sum_of_ur([]), 0)
expect(sum_of_ur(LOF1), 30.5)
expect(sum_of_ur(LOF2), 65.8)
expect(sum_of_ur([19.3, 11.2, 11.0, 18.5, 5.8, 9.1, 5.3, 35.0, 17.8]), 133.0)


# Examples and tests for num_of_ur
expect(num_of_ur([]), 0)
expect(num_of_ur(LOF1), 2)
expect(num_of_ur(LOF2), 5)
expect(num_of_ur([19.3, 11.2, 11.0, 18.5, 5.8, 9.1, 5.3, 35.0, 17.8]), 9)


# Examples and tests for avg_ur
expect(avg_ur([]),0)
expect(avg_ur([19.3, 11.2, 11.0, 18.5]), 15.0)
expect(avg_ur(LOF1), 15.25)
expect(avg_ur(LOF2), 13.16)


# Examples and tests for avg_ur_of_ur_list
expect(avg_ur_of_ur_list([]), 0)
expect(avg_ur_of_ur_list(LOCD1), 15.25)
expect(avg_ur_of_ur_list(LOCD2), 13.16)
expect(avg_ur_of_ur_list([CD2, CD3]), 14.75)
expect(avg_ur_of_ur_list([CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8]), 16.375)


# Examples and tests for is_greater
expect(is_greater(3.0, 3.0), False)
expect(is_greater(7.3, 2.9), True)
expect(is_greater(9.9, 8.8), True)
expect(is_greater(4.7, 8.4), False)
expect(is_greater(5.4, 11.7), False)


# Examples and tests for is_greater_than_avg
expect(is_greater_than_avg(CD0, LOCD2), True)
expect(is_greater_than_avg(CD0, LOCD1), True)
expect(is_greater_than_avg(CD1, LOCD1), False)
expect(is_greater_than_avg(CD2, [CD0,CD1,CD2]), False)
expect(is_greater_than_avg(CD3, [CD0,CD1,CD2,CD3]), True)
expect(is_greater_than_avg(CD3, LOCD2), True)
expect(is_greater_than_avg(CD3, [CD3]), False)
expect(is_greater_than_avg(CD4, LOCD2), False)
expect(is_greater_than_avg(CD9, [CD0,CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8,CD9]), True)

# Examples and tests for filter_the_data
expect(filter_the_data([]), [])
expect(filter_the_data(LOCD1), [CD0])
expect(filter_the_data(LOCD2), [CD0,CD3])
expect(filter_the_data([CD0,CD1,CD2]), [CD0])
expect(filter_the_data([CD0,CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8,CD9]), [CD6, CD7, CD9])

# Examples and tests for plot_gdp_over_ur
# a plot for an empty list (no point shows on the plot)
expect(plot_gdp_over_ur([]), None) 

# a plot for a list of 9 CountryData (5 points show on the plot)
expect(plot_gdp_over_ur([CD0,CD1,CD2,CD3,CD4,CD5,CD6,CD7,CD8]), None)

# a plot for the list LOCD2 (2 points show on the plot)
expect(plot_gdp_over_ur(LOCD2), None)



summary()
```
